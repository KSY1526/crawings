```python
import pandas as pd
import re
```

```python
path = './save_csv/'

lsts = []

for i in range(1, 3):
    lst = pd.read_csv(f'{path}river_behind{i}.csv')
    lsts.append(lst)
```

```python
data = lsts[0]

for i in range(1, 2):
    data = pd.concat([data,lsts[i]])

data
```

```python
data['url']= data['url'].apply(lambda x: re.findall(r'/[0-9]+', x)[0][1:])
data['review'] = data['review'].apply(lambda x: re.findall(r'[\d{1.2}]+', x))
data
```

```python
data = data.drop_duplicates('restaurant') # restaurant
```
